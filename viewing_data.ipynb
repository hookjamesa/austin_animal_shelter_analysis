```python
# Importing dependencies.
import pandas as pd
```

```python
# Read in files.
file_path = "Resources/aac_intakes_outcomes.csv"
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
# Checking for nan
df.count()
```

```python
# Dropping nan
df.dropna(inplace=True)
```

```python
# Analyzing remaining data
df.count()
```

```python
# Selecting columns.
df_samp = df.filter(["animal_id_outcome", "outcome_subtype", "outcome_type", "sex_upon_outcomes", 
            "age_upon_outcome_(years)", "age_upon_outcome_age_group", "outcome_monthyear", "age_upon_intake", 
            "animal_id_intake", "animal_type", "breed", "color", "intake_condition", "age_upon_intake_(years)", 
            "intake_year", "time_in_shelter_days"], axis=1)
```

```python
# Get dummy variables
dummy = pd.get_dummies(df_samp["outcome_type"])
```

```python
## Adding dummy variable back into dataframe
df_2 = pd.concat((df_samp, dummy), axis=1)
```

```python
## outcome_type 
df_2["outcome_type"].unique()
```

```python
# Dropping Extra Columns
df_drop = df_2.drop(['outcome_type', 'Died', 'Euthanasia', 'Transfer', 'Missing', "Return to Owner"], axis=1)
```

```python
# Sample for dataframe.
aac_sample = df_drop.sample(n=100)
```

```python
# Ensuring sample was taken
aac_sample.count()
```

```python
# New dataframe types.
aac_sample.dtypes
```

```python
## View dataframe
aac_sample.head()
```

```python
### Extracing file.
aac_sample.to_csv('Resources/aac_sample.csv', index=True)
```
